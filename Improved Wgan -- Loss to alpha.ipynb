```python
def plot_scatter_gp0(iter):
    data = np.load('gp0/alpha_to_disc_cost_wgan_gp0_%d.npy' % iter)
    #data_avg = data[i]
    #print data_avg.shape
    
    f, ax1 = plt.subplots(1, 1, sharey=True)
    f.set_facecolor('white')
    
    #fig = plt.figure()
    #ax1 = fig.add_subplot(111)
    ax1.set_xlabel('diff')
    ax1.set_ylabel('depth')
    ax1.set_title('diff-depth gp0  Iter: %d' % iter)
    
    
    data_avg = data
    st = data_avg[:,0]
    mid = data_avg[:,50]
    en = data_avg[:,99]
    dif = en-st
    dep = (st+en)/2 - mid
    
    ax1.scatter(dif, dep)
    plt.plot()
```
